### **4. Findings**

#### 4.1 Clustering
- The model with 2 clusters provides the best separation between clusters, as indicated by the highest silhouette score and the lowest Davies-Bouldin index.

- As the number of clusters increases, the silhouette score decreases, indicating less distinct separation between clusters.

- The Davies-Bouldin index also increases with the number of clusters, suggesting that the clustering quality deteriorates as the number of clusters increases.

- Cluster near the origin represents flights that were on time or had minimal delays for both departure and arrival. The larger this cluster, the more consistent the airline is at operating on schedule.

- Lower left clusters represent flights with delays in both departure and arrival. The distance of the centroid from the origin indicates the severity of the delays. A cluster closer to the axes suggests minor delays, while one further indicates significant delays for both.

#### 4.3 Chi-Square Test

The Chi-Square Test of Independence is used to determine whether there is a significant association between two categorical variables. In this case, the test is applied to each pair of a categorical variable and the 'cluster' variable.

- Gender_code: The Chi-Square value is 0.93 with a p-value of 0.82. This indicates that there is no significant association between gender and the clusters.

- Customer Type_code: The Chi-Square value is 1.46 with a p-value of 0.69. Similar to gender, there is no significant association between customer type and the clusters.

- Type of Travel_code: The Chi-Square value is 1.08 with a p-value of 0.78. Again, there is no significant association between the type of travel and the clusters.

- Class_code: The Chi-Square value is 2.38 with a p-value of 0.88. There is no significant association between travel class and the clusters.

- Satisfaction_code: The Chi-Square value is 0.50 with a p-value of 0.92. There is no significant association between satisfaction level and the clusters.

- Other variables: Similar interpretations can be drawn for other variables. None of them show a significant association with the clusters.

#### 4.4 ANOVA

Departure Delay in Minutes and Arrival Delay in Minutes:

- The F-values for both variables are extremely high, indicating significant differences in departure and arrival delays among the clusters.

- The p-values for both variables are 0.0, indicating strong evidence against the null hypothesis that there are no differences in departure and arrival delays among the clusters.

Departure Delay in Minutes_mmnorm and Arrival Delay in Minutes_mmnorm:

- The F-values for both variables are relatively low, suggesting that there are no significant differences in the standardized (min-max normalized) departure and arrival delays among the clusters.

```python
#K-Means Clustering


# 2.1.1. Determine Value of 'K' in K-Means using Elbow Curve & KMeans-Inertia

'''
KMeans-Inertia : Sum of Squared Distances of Samples to their closest Cluster Center (Centroid), Weighted by the Sample Weights (if provided)
'''
train_new_df=df_ppd[['Gender_code', 'Customer Type_code', 'Type of Travel_code', 'Class_code', 'satisfaction_code',
                         'Inflight wifi service_code', 'Departure/Arrival time convenient_code', 'Ease of Online booking_code',
                         'Gate location_code', 'Food and drink_code', 'Online boarding_code', 'Seat comfort_code',
                         'Inflight entertainment_code', 'On-board service_code', 'Leg room service_code',
                         'Baggage handling_code', 'Checkin service_code', 'Inflight service_code', 'Cleanliness_code','Departure Delay in Minutes', 'Arrival Delay in Minutes',  'Departure Delay in Minutes_mmnorm', 'Arrival Delay in Minutes_mmnorm']]
wcssd = [] # Within-Cluster-Sum-Squared-Distance
nr_clus = range(1,11) # Number of Clusters
for k in nr_clus:
    kmeans = kmclus(n_clusters=k, init='random', random_state=111)
    kmeans.fit(train_new_df)
    wcssd.append(kmeans.inertia_)
plt.plot(nr_clus, wcssd, marker='x')
plt.xlabel('Values of K')
plt.ylabel('Within Cluster Sum Squared Distance')
plt.title('Elbow Curve for Optimal K')
print(plt.show())


cluster_range = [2, 3, 4, 5]

# Iterate over different numbers of clusters
for k in cluster_range:
  # Perform Kmeans Clustering
    km_cluster = kmclus(n_clusters=k, init='random', random_state=222)
    print(km_cluster)
    km_cluster_model = km_cluster.fit_predict(train_new_df);
    print(km_cluster_model)
# Calculate Silhouette Score
    sscore_km_cluster = sscore(train_new_df, km_cluster_model);
    print(f"Silhouette Score for {k} clusters: {sscore_km_cluster}")
    # Calculate Davies-Bouldin Index
    dbscore_km_cluster = dbscore(train_new_df, km_cluster_model)
    print(f"Davies-Bouldin Index for {k} clusters: {dbscore_km_cluster}\n")
```

```python
import time
import psutil
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns

# Start time
start_time = time.time()

# Specify the number of clusters
num_clusters = 4

# Initialize KMeans object
kmeans = KMeans(n_clusters=num_clusters, random_state=42)

# Fit KMeans clustering model to the data
kmeans.fit(train_new_df[['Gender_code', 'Customer Type_code', 'Type of Travel_code', 'Class_code', 'satisfaction_code',
                         'Inflight wifi service_code', 'Departure/Arrival time convenient_code', 'Ease of Online booking_code',
                         'Gate location_code', 'Food and drink_code', 'Online boarding_code', 'Seat comfort_code',
                         'Inflight entertainment_code', 'On-board service_code', 'Leg room service_code',
                         'Baggage handling_code', 'Checkin service_code', 'Inflight service_code', 'Cleanliness_code','Departure Delay in Minutes', 'Arrival Delay in Minutes',  'Departure Delay in Minutes_mmnorm', 'Arrival Delay in Minutes_mmnorm']])

# End time
end_time = time.time()

# Total time taken
time_taken = end_time - start_time
print("Time taken by K-Means for k=4:", "{:.4f} seconds".format(time_taken))

# Memory used by the KMeans model
memory_used = psutil.Process().memory_info().rss / 1024 / 1024  # Memory in MB
print("Memory used by K-Means for k=4:", "{:.2f} MB".format(memory_used))

# Get cluster labels
cluster_labels = kmeans.labels_

# Add cluster labels to DataFrame
train_new_df['cluster'] = cluster_labels

# Get centroids of each cluster
centroids = kmeans.cluster_centers_

# Visualize the clusters using seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(data=train_new_df, x='Departure Delay in Minutes', y='Arrival Delay in Minutes', hue='cluster', palette='viridis', s=100)
plt.scatter(centroids[:, 5], centroids[:, 6], marker='*', s=300, c='black', label='Centroids')
plt.xlabel('Departure delay in Minutes')
plt.ylabel('Arrival delay in Minutes')
plt.title('K-means Clustering with Centroids')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
# Further analysis or visualization with cluster labels

# Cluster Analysis with Categorical Variables: Chi-Square Test of Independence
# Assuming you have performed clustering and added cluster labels to your DataFrame 'df'

import pandas as pd
from scipy.stats import chi2_contingency

# Select categorical variables for analysis
categorical_variables = ['Gender_code', 'Customer Type_code', 'Type of Travel_code', 'Class_code', 'satisfaction_code',
                         'Inflight wifi service_code', 'Departure/Arrival time convenient_code', 'Ease of Online booking_code',
                         'Gate location_code', 'Food and drink_code', 'Online boarding_code', 'Seat comfort_code',
                         'Inflight entertainment_code', 'On-board service_code', 'Leg room service_code',
                         'Baggage handling_code', 'Checkin service_code', 'Inflight service_code', 'Cleanliness_code', 'cluster']

# Create a contingency table for each categorical variable and the cluster labels
contingency_tables = {}
for variable in categorical_variables:
    contingency_tables[variable] = pd.crosstab(train_new_df[variable], train_new_df['cluster'])

# Perform Chi-Square Test of Independence for each contingency table
chi2_results = {}
for variable, contingency_table in contingency_tables.items():
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    chi2_results[variable] = {'Chi-Square': chi2, 'P-value': p}

# Display the results
for variable, result in chi2_results.items():
    print(f"Chi-Square Test of Independence for variable '{variable}':")
    print(f"   - Chi-Square Value: {result['Chi-Square']}")
    print(f"   - P-value: {result['P-value']}")


```

```python
from scipy.stats import f_oneway

# List of features for ANOVA
features = ['Departure Delay in Minutes', 'Arrival Delay in Minutes',  'Departure Delay in Minutes_mmnorm', 'Arrival Delay in Minutes_mmnorm']

# Initialize an empty dictionary to store ANOVA results
anova_results = {}

# Iterate over unique cluster labels
for cluster_label in train_new_df['cluster'].unique():
    # Filter dataframe for the current cluster
    cluster_df = train_new_df[train_new_df['cluster'] == cluster_label]

    # Initialize an empty dictionary to store ANOVA results for the current cluster
    cluster_anova_results = {}

    # Perform ANOVA for each feature
    for feature in features:
        # Perform ANOVA test
        anova_result = f_oneway(*[train_new_df[train_new_df['cluster'] == i][feature] for i in train_new_df['cluster'].unique()])
        # Store ANOVA result
        cluster_anova_results[feature] = anova_result

    # Store ANOVA results for the current cluster
    anova_results[cluster_label] = cluster_anova_results

# Print ANOVA results
for cluster_label, cluster_anova_result in anova_results.items():
    print(f"Cluster {cluster_label} ANOVA Results:")
    for feature, result in cluster_anova_result.items():
        print(f"{feature}: F-value = {result.statistic}, p-value = {result.pvalue}")
```
